```python
#warna yang akan digunakan
RED1, RED2 = '#800000', '#880808'
GRAY1, GRAY2, GRAY3, GRAY4, GRAY5 = '#555655', '#646369', '#828282', '#929497', '#A6A6A5'
```

```python
X = ['Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3']
Ymerah = np.array([4, 3, 5, 4, 5, 5, 12, 15, 20, 33, 42]) 
Ytengah = np.array([54, 62, 70, 71, 75, 60, 58, 60, 70, 64, 48]) 
Yatas = np.array([42, 35, 25, 25, 25, 35, 30, 25, 10, 3, 10]) 
```

```python
#buat gambar untuk membuat visualisasi
fig, ax1 = plt.subplots(figsize=(11.5, 7.2),  #mengatur pXl
                        dpi=110)  
fig.subplots_adjust(left=0.103, right=0.7, top=0.881, bottom=0.096)

#mengatur limit sumbu x dan y
ax1.set_ylim([-15, 100])

#memindahkan sumbu x ke 0
ax1.spines['bottom'].set_position('zero')

# pengaturan tulisan/titik pada sumbu
plt.setp(ax1,
         xticks=list(range(11)),
         xticklabels=X,
         yticks=list(range(0, 110, 10)))

#buat grafik dan atur sesuai dengan nilai yang telah ditentukan
white_bars = plt.bar(list(range(0, 11)), Ymerah, width=0.70, color=RED1, 
        edgecolor=RED1, linewidth=0.9)
plt.bar(list(range(0, 11)), Ytengah, width=0.70, bottom=Ymerah, color=GRAY5, 
        edgecolor=GRAY5, linewidth=0.9)
plt.bar(list(range(0, 11)), Yatas, width=0.70, bottom=Ymerah + Ytengah, color=GRAY3, 
        edgecolor=GRAY3, linewidth=0.9)

#tambahkan data pada grafik yang berwarna merah
for i, b in enumerate(white_bars):
    if Ymerah[i] <= 10:
        continue
    x = b.get_x() + b.get_width() / 2
    y = b.get_y() + b.get_height() - 4
    ax1.text(x, y,
             str(Ymerah[i]) + '%',
             fontsize=12,
             horizontalalignment='center',
             color='white')

#hapus border pada sisi atas dan kanan
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_color(GRAY5)
ax1.spines['left'].set_color(GRAY5)
ax1.spines['left'].set_linewidth(1.5)
ax1.spines['bottom'].set_linewidth(1.5)

#ubah warna sumbu Y menjadi abu-abu
ax1.tick_params(axis='y', colors=GRAY1)

#atur warna dan ukuran di sumbu x
for item in ax1.get_xticklabels():
    item.set_fontsize(10)
    item.set_color(GRAY2)

#buat judul di sumbu y
ax1.text(-1.90, 75, '% of total projects', fontsize=10, rotation='vertical', 
         color=GRAY1)

# Tambahkan informasi dalam bentuk tulisan
ax1.text(4.83, 108, 'As of Q3 2015,', fontsize=14, color=GRAY2, linespacing=1)
ax1.text(4.80, 104,
         '                        $\\bf{more\ than\ 1/3\ of}$'
         '\n         '
         '$\\bf{project\ are\ missing\ goals}$',
         fontsize=14, color=RED2, linespacing=1)

#berikan tulisan di bawah gambar mengenai sumber data
ax1.text(-1.84, -18,
         'Data source: XYZ Dashboard; the total number of projects has increased over time from 230 in early 2013 to nearly 270 in Q3 2015.',
         fontsize=8.3, color=GRAY4)

ax1.text(0, -12, '            2013                                   2014                                2015', fontsize=12, color=GRAY1)

#judul
ax1.text(-2, 114, 'Goal attainment over time', 
         fontsize=16, color=GRAY1)

ax1.legend(['Miss', 'Meet','Exceed'], loc='upper right', bbox_to_anchor=(0.28, 1.11), ncol=3, fontsize=8)
```
